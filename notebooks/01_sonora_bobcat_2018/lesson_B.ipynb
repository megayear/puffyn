```python
from gollum.sonora import SonoraSpectrum
import numpy as np
import matplotlib.pyplot as plt
from specutils import Spectrum1D
import pandas as pd
import astropy.units as u
%config InlineBackend.figure_format='retina'
```

## Read-in the synthetic data

```python
df = pd.read_csv('../../data/synthetic/01_sonora_bobcat_2018/lesson_B-specimen01.csv')
```

```python
df.head()
```

```python
spectrum = Spectrum1D(spectral_axis=df.wavelength_Angstroms.values*u.Angstrom, 
                           flux=df.flux.values*u.ct)

wl_lo, wl_hi = (spectrum.wavelength.value.min(), 
                spectrum.wavelength.value.max())
```

```python
plt.figure(figsize=(14, 4))
plt.step(spectrum.wavelength, spectrum.flux)
plt.xlabel('$\lambda \;(\AA)$'); plt.ylabel('Flux');
plt.ylim(0, 2.0);
```
